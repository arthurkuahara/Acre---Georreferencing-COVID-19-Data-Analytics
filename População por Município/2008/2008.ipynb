```python
import geopandas as gpd
import pandas as pd
import matplotlib
```

```python
pop08 = pd.read_csv("2008.csv",sep = ";")
pop08["GEOCODIG_M"] = pop08["COD_UF"]*100000 + pop08["COD_MN"]
pop08
```

```python
br08 = gpd.read_file("shp/55mu2500psd.shp")
br08
```

```python
mapa = br08.merge(pop08, on = "GEOCODIG_M")
mapa.plot(column = "POP_ESTIM", figsize = (20,20), 
          legend = True, legend_kwds={'label': "População Estimada em 2008", 'orientation': "horizontal"})
```
